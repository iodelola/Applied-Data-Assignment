```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.express as px


# reading all csv's
df_access_electricity = pd.read_csv("Access_electricity.csv")
df_Electri_consumption= pd.read_csv("Electric_power_consumption.csv")
```

```python
df_access_electricity.head()
```

```python
# add columns with Elements
df_access_electricity["Element"] = "Access to Electricity"
df_Electri_consumption["Element"] = "Electric Power consumption"


df_Electri_consumption.head()
```

```python
#merge both csv files. 
df_climate_change = pd.concat([df_access_electricity, df_Electri_consumption])
df_climate_change.head()
```

```python
# sort and save as csv file
df_climate_change = df_climate_change.sort_values(["Country Name"])
df_climate_change.to_csv("climate_change_info.csv")
```

```python
# statistical overview
df_climate_change.describe()
```

```python
# Line charts of dataset, Access to electricity and Electric power consumption of the whole world.
df_elements = df_climate_change.groupby("Element").sum()
df_elements
```

```python
#Transpose the dataframe
elements_t = df_elements.T
elements_t.head()
```

```python
plt.scatter(elements_t['Access to Electricity'], elements_t['Electric Power consumption'])
```

```python
#Pot a line chat for the Indicator
sns.set_style("darkgrid")
line = sns.lineplot(x=elements_t.index, 
                    y=elements_t['Access to Electricity'], 
                    data=elements_t, color="red", marker='o')
plt.xticks(rotation=45)
figsize=(18, 8)
plt.tight_layout()
```

```python
#line plot
sns.set_style("darkgrid")
line = sns.lineplot(x=elements_t.index, 
                    y=elements_t['Electric Power consumption'], 
                    data=elements_t, color="blue", marker='+')
plt.xticks(rotation=45)
figsize=(50, 8)
plt.tight_layout()
```

```python
#correlation matrices
df_climate_change.corr()
```

```python
#Diplay an Heatmap,graphical representation of data where values are depicted by color.
f,ax = plt.subplots(figsize=(10, 10))
sns.heatmap(df_climate_change.corr(), annot=True, linewidths=.8, fmt= '.1f',ax=ax)
```

```python
#correlation table 
df_climate_change.corr(method="kendall")
```

```python
#save excel files
df_climate_change.to_excel("pearson.xlsx")
df_climate_change.to_excel("kendall.xlsx")
```
